**DFS**
_Deep First Search_


```python
G = [[1, 3], 
     [0, 3, 2], 
     [1, 4], 
     [0, 1],
     [2]]

used = [0] * 5
```

```python
def dfs(cur):
    used[cur] = 2 # отметка о посещении вершины и начале обработки

    print(cur)

    for to in G[cur]:
        if used[to] == 2:
            # цикл
            pass
        if not used[to]:
            dfs(to)
        

    used[cur] = 1 #отметка об окончании обработки
```

```python
dfs(0)
```
