```python
def build_tree(l_b, r_b, i):
    if (r_b - l_b == 1):
        tree[i] = arr[l_b]
        print(i, tree[i])
        return tree[i]

    m_b = (l_b + r_b) // 2
    
    tree[i] = max(
        build_tree(l_b, m_b, i * 2 + 1),
        build_tree(m_b, r_b, i * 2 + 2)
    )
    print(i, tree[i])
    return tree[i]


build_tree(0, len(arr), 0)
```

```python
def get_max(c_l, c_r, l_b, r_b, i):
    if (c_l >= l_b and c_r <= r_b):
        return tree[i]
    if (c_l >= r_b or c_r <= l_b):
        return 0
    
    m_b = (c_l + c_r) // 2
    
    return max(
        get_max(c_l, m_b,  l_b, r_b, i * 2 + 1),
        get_max(m_b, c_r,  l_b, r_b, i * 2 + 2)
    )
    
```

```python
def upd_one_elem(l, r, pos, new_val, i = 0):
    if (r - l == 1):
        if (l == pos):
            tree[i] = new_val
            return
        else:
            return
        
    if (pos >= l and pos < r):
        m = (l + r) // 2
        upd_one_elem(l ,m, pos, new_val, i * 2 + 1)
        upd_one_elem(m, r, pos, new_val, 2*i + 2)
        tree[i] = max(tree[2*i + 1], tree[2*i + 2])    
```

```python
tree_push = [0] * len(tree) #пусть 0 - не может быть величиной на котороую заменяем все эллементы отрезка
```

```python
def relax(l, r, i):
    if (r - l == 1):
        tree[i] = tree_push[i]
        return
    tree_push[2*i + 1] = tree_push[i]
    tree_push[2*i + 2] = tree_push[i]
```

```python
def get_max_node(i):
    if (tree_push[i]):
        return tree_push[i]
    else:
        return tree[i]
```

```python
def upd_int(c_l, c_r, pos_l, pos_r, new_val, i = 0):
    if (c_l >= pos_r or c_r <= pos_l):
        return
    if (tree_push[i] != 0):
        relax()
    if (c_r <= pos_r and c_l >= pos_l):
        tree[i]         = max(new_val, tree[i])
        tree_push[i]    = new_val
    else:
        m = (c_l + c_r) // 2
        upd_int(c_l, m, pos_l, pos_r, new_val, 2 * i + 1)
        upd_int(m, c_r, pos_l, pos_r, new_val, 2 * i + 2)
        tree[i] = max(
            get_max_node(2 * i + 1),
            get_max_node(2 * i + 2)
        )
    
```

```python
upd_int(0, len(arr), 3, 7, 200)
```
